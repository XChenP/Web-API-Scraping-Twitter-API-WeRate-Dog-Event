**Dog Stages**

There are 351 Tweeter users gave inforamtion on their dog stages (definitions are based on those given by Dogtionary). Below is a breakdown of the stages. About 64%(225/351) of dogs are pupper while no dog is snoot. 

```python

import seaborn as sns 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
count = [75, 225, 28, 1, 0, 22]
stage = ['Doggo', 'Pupper', 'Puppo', 'Blep', 'Snoot', 'Floof']
y_pos = np.arange(len(stage))
plt.bar(range(len(stage)), count,  color = (0.5,0.1,0.5,0.6))
plt.xticks(range(len(stage)), stage)
plt.xlabel('Dog Stage')
plt.ylabel('Count')
plt.title('Dog Stage Distribution')
plt.show()
```

**Tweet Day VS Retweet Counts**
<br>
Is there a relationship between the day of week when someone tweeted a dog picture and the retweet counts it got? From the scatterplot below, it doesn't seem to have a trend where tweet day affect retweet count. Our linear regression model also confirm that. 

```python

from sklearn.linear_model import LinearRegression
df=pd.read_csv('twitter_archive_master.csv')
plt.scatter(x= df['Weekday'], y = df['Retweet_Count'])
plt.xlabel('Weekday')
plt.ylabel('Retweet Count')
plt.title('Weekday and Retweet Count')
plt.show()
```

```python

x = df['Weekday']
y = df['Retweet_Count']
model = LinearRegression(fit_intercept=True)
model.fit(x[:, np.newaxis], y)
xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit)
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```
